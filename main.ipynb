```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
FILENAME = input("Please enter a filename (.csv): ")
try:
    df = pd.read_csv(FILENAME)
except FileNotFoundError:
    print("File not found. Please try again.")
    exit()
except Exception as e:
    print(f"There's an error {e}. Please try again.")
    exit()
print(df.head())
```

```python
df = pd.read_csv(FILENAME) 

x_col = input("What's your independent variable (X)?: ")
y_col = input("What's your dependent variable (Y)?: ")

try:
    X = df[[x_col]]
    Y = df[[y_col]]
except KeyError:
    print("One or both column names do not exist in the file.")
```

```python
plt.scatter(df[x_col], df[y_col])
```

```python
lr_model = LinearRegression()
lr_model.fit(X, Y)
```

```python
y_predict = lr_model.predict(X)
plt.scatter(X, Y, color='skyblue', label='Data points')
plt.plot(X, y_predict, color='Red', linewidth=2, label='Regression line')

plt.title('Linear Regression Visualization')
plt.xlabel(x_col)
plt.ylabel(y_col)
```

```python
def predict_value():
    while True:
        try:
            value = float(input(f"Enter a value of '{x_col}' to predict '{y_col}' (or type 'exit' to quit): "))
            prediction = lr_model.predict([[value]])
            print(f"Predicted value of {y_col} for {x_col} = {value} is: {prediction[0][0]:.2f}")
        except ValueError:
            print("Exiting prediction tool.")
            break
predict_value()
```
